Calculation for the average number of plays each team makes per game since 2009

```python
# Clean up Jacksonville Data to be JAC
r.rs_pbp.posteam = r.rs_pbp.posteam.apply(lambda x: 'JAC' if x == 'JAX' else x)
# Clean up LA Rams/ STL Rams to just be STL Rams
r.rs_pbp.posteam = r.rs_pbp.posteam.apply(lambda x: 'STL' if x == 'LA' else x)
play_count_pt = r.rs_pbp.loc[~r.rs_pbp.down.isnull()].pivot_table(values='Date', index='posteam', columns='Season', aggfunc='count')
play_count_pt['mean'] = play_count_pt.mean(axis=1)
play_count_pt.loc['MEAN'] = play_count_pt.mean()
play_count_pt/16
```
